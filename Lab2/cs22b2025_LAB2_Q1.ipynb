```python
# a) Create a custom function that calculates BMI (Body Mass Index) using the formula: BMI = weight(kg) / (height(m))^2. Apply this function to create a new 'BMI' column. (Assume 'Weight' is in kg and 'Height' is in meters)
def calc_BMI():
    weight = int(input("Enter weight in kg:"))
    height = int(input("Enter height in meters:"))
    bmi = weight/height**2
    return bmi
```
