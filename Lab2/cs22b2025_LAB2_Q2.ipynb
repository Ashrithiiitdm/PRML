```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt

# For x-values in all plots
x = np.linspace(0, 10, 100)
#Create a figure of required size
plt.figure(figsize = (12, 10))

# Creating the 2X2 subplots
plt.subplot(2, 2, 1)
plt.title('Plot 1')
plt.subplot(2, 2, 2)
plt.title('Plot 2')
plt.subplot(2, 2, 3)
plt.title('Plot 3')
plt.subplot(2, 2, 4)
plt.title('Plot 4')

#Showing the plot
plt.show()
```

```python
 
plt.figure(figsize = (12, 10))

# Creating the 2X2 subplots
plt.subplot(2, 2, 1)
plt.title('Plot 1')
# Defining the lines
y1 = x
y2 = x**2
y3 = x**3
# a), b), c) Plotting the lines using different colors and line widths
plt.plot(x, y1, label = 'y = x', linewidth = 2, color = 'red')
plt.plot(x, y2, label = 'y = x^2', linewidth = 4, color = 'blue')
plt.plot(x, y3, label = 'y = x^3', linewidth = 6, color = 'green')
# e) Title for plot
plt.title("Basic Line Plots")

# d) Adding a legend
plt.legend()
#Remaining plots
plt.subplot(2, 2, 2)
plt.title('Plot 2')
plt.subplot(2, 2, 3)
plt.title('Plot 3')
plt.subplot(2, 2, 4)
plt.title('Plot 4')

#Showing the plot
plt.show()
```

```python
 
plt.figure(figsize = (12, 10))

# Creating the 2X2 subplots
plot1 = plt.subplot(2, 2, 1)
plt.title('Plot 1')
# Defining the lines
y1 = x
y2 = x**2
y3 = x**3
#Plotting the lines using different colors and line widths
plt.plot(x, y1, label = 'y = x', linewidth = 2, color = 'red')
plt.plot(x, y2, label = 'y = x^2', linewidth = 4, color = 'blue')
plt.plot(x, y3, label = 'y = x^3', linewidth = 6, color = 'green')
#Title for plot
plt.title("Basic Line Plots")
#Labels for X and Y axis
plt.xlabel('x')
plt.ylabel('y')
#Adding a legend
plt.legend()
#  3rd question starts from here
y21 = np.sin(x)
plt.subplot(2, 2, 2)
plt.title('Plot 2')
# a), b), c) Plotting the line with different line styles and same color
# Solid Line
plt.plot(x, y21, label = 'y = sinx', color = 'black', linestyle = '-')
# Dashed
plt.plot(x, y21, label = 'y = sinx', color = 'black', linestyle = '--')
# Dashed-dot
plt.plot(x, y21, label = 'y = sinx', color = 'black', linestyle = '-.')
# Dotted
plt.plot(x, y21, label = 'y = sinx', color = 'black', linestyle = ':')
# e) Title
plt.title("Line Styles")
# d) Added legend 
plt.legend()

#4th question starts from here
y31  = np.cos(x)
plt.subplot(2, 2, 3)
# a), b), c) , d)
plt.plot(x, y31, label = 'y = cosx', color = 'blue', linestyle = '-', marker = '+', markersize = 5)
plt.plot(x, y31, label = 'y = cosx', color = 'blue', linestyle = '-', marker = 'o', markersize = 8)
plt.plot(x, y31, label = 'y = cosx', color = 'blue', linestyle = '-', marker = 's', markersize = 11)
plt.plot(x, y31, label = 'y = cosx', color = 'blue', linestyle = '-', marker = '^', markersize = 14)
# f) Title
plt.title('Markers')
# e) Adding a legend
plt.legend()

# 5th question starts here
# tanx line
y41 = np.tan(x)
plt.subplot(2, 2, 4)

# a, c, d
line, = plt.plot(x, y41, label = 'y = tanx', color = 'black', marker = 'o', markerfacecolor = 'yellow', markeredgecolor = 'red')
# b here line holds the Line2D object returned by plt.plot
line.set_dashes([5, 5])
# e Title
plt.title('Custom Line and Marker Styles')
# Legend
plt.legend()

#Showing the plot
plt.show()
```
